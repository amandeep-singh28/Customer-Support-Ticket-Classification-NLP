### 1. Analyzing Prediction DataFrame

To analyze the model's performance, the actual labels, predicted labels, and corresponding ticket descriptions are combined into a single DataFrame.

This makes it easier to identify correctly classified and misclassified tickets.

```python
import pandas as pd
pd.set_option("display.max_colwidth", None)
```

```python
results = pd.read_csv('baselineResults.csv')
```

```python
results.head()
```

```python
correct_predictions = results[
    results["Actual"] == results["Predicted"]
]

print(f"Correct Predictions : {len(correct_predictions)}")
```

```python
wrong_predictions = results[
    results["Actual"] != results["Predicted"]
]

print(f"Wrong Predictions : {len(wrong_predictions)}")
```

```python
misclassification_rate = (
    len(wrong_predictions) / len(results)
) * 100

print(f"Misclassification Rate : {misclassification_rate:.2f}%")
```
